```python
data.isna().sum()
zoneless = pd.isnull(data["Region"]) 
data[zoneless]
data["Region"].fillna("Unspecified", inplace=True) 

Region = data['year'].unique()
len(Region)

for i in range(5):
    x = data[data['year'] == Region[i]]
    plt.scatter(x['Economy_GDP_per_Capita'],x['Happiness_Rank'])
    plt.xlabel("Economy_GDP_per_Capita")
    plt.ylabel("Happiness_Rank")
plt.show()
```

```python
boxplot_data= data.drop(['Standard_Error','Family', 'Trust(Government_Corruption)','Dystopia_Residual','Lower_Confidence_Interval','Upper_Confidence_Interval','Whisker.high','Whisker.low','Social support'],axis=1)

group=np.repeat(np.array(['2015', '2016','2017', '2018','2019']), [158,157,155,156,156], axis=0).tolist()
data=boxplot_data.copy()
data['year']= group

props = dict(whiskers="black", medians="red", caps="black")
data.groupby(by='year').boxplot(column='Economy_GDP_per_Capita',patch_artist=True,
                               color=props,layout=(2,10), fontsize=15, figsize=(40,20));
```
