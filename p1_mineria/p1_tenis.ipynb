```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import warnings
import seaborn as sns
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('MEN_SHOES.csv')
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data['Current_Price'] = data['Current_Price'].str.replace('â‚¹', '')
data['Current_Price'] = data['Current_Price'].str.replace(',', '')
data['Current_Price'].unique()
data.dropna(axis = 0, inplace = True)
data['Current_Price'] = data['Current_Price'].astype(int)

data['How_Many_Sold'] = data['How_Many_Sold'].str.replace(',', '')
data['How_Many_Sold'].unique()
data.dropna(axis = 0, inplace = True)
data['How_Many_Sold'] = data['How_Many_Sold'].astype(int)
```

```python
data.describe()
```

```python
plt.figure(figsize = (20,10))
sns.barplot(x = 'Brand_Name',
            y = 'RATING',            
            data = data).set(title = "Rating vs Brand Name");
plt.xticks(rotation = 90)
plt.show()
```

```python
data = data.drop_duplicates()

colors = {'ASIAN': 'Blue',
          'Adidas': 'Red',
          'Axter': 'Green',
          'BATA': 'Yellow',
          'Bourge': 'Purple',
          'Campus': 'Orange',
          'Centrino': 'Pink',
          'Corstyle': 'Brown',
          'D Shoes': 'Cyan',
          'FEETEES': 'Magenta',
          'FURO': 'Teal',
          'Generic': 'Lime',
          'Kraasa': 'Gray',
          'Nivia': 'Olive',
          'Reebok': 'Maroon',
          'Robbie jones': 'Navy',
          'Sparx': 'Aqua',
          'URJO': 'Fuchsia',
          'Wakefield': 'Silver',
          'road runner': 'Black'
         }
brands_color = data.Brand_Name.map(colors)
```

```python
fig, ax = plt.subplots()
for brands in set(data.Brand_Name):
    ax.scatter(
        data.RATING[data.Brand_Name == brands],
        data.Current_Price[data.Brand_Name == brands],
        s = 30,
        c = colors[brands],
        label = brands)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.xlabel("RATING")
plt.ylabel("Current_Prices")
plt.show()
```
