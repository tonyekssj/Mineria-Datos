```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv('listings.csv')
data.shape
```

```python
missing_values_count=data.isnull().sum()
data.isnull().sum()
```

```python
total_datos_perdidos = missing_values_count.sum()
total_datos_perdidos
```

```python
data=data.dropna(subset=['name'])
data.isnull().sum()
```

```python
data['last_review'].mode
```

```python
data['last_review'].fillna(data['last_review'].mode()[0],inplace = True)
data.isnull().sum()
```

```python
data['reviews_per_month'].mode
```

```python
data
```

```python
data['reviews_per_month'] = data['reviews_per_month'].fillna(data['reviews_per_month'].mean())
data.isnull().sum()
```
